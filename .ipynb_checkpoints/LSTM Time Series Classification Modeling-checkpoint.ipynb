#### Functions for Feature Engineering

```python
def transformed_features(x, y, T):
    # scaling the features
    sc = StandardScaler()
    x_std = pd.DataFrame(sc.fit_transform(x.values),
                               index=x.index,
                               columns=x.columns)
    x_std, y_std= TimeSeries_data(x_std, y, T)
    return x_std, y_std
```

```python
def TimeSeries_data (x, y, T):
    inputs, target = [], []
    for i in range(y.shape[0] - (T)):
        inputs.append(x.iloc[i:i+T].values)
        target.append(y.iloc[i + (T)])
    inputs, target = np.array(inputs), np.array(target).reshape(-1,1)
    return inputs, target
```

```python
def lstm_transformed_data(features, target, Year_column, T):
    cor_matrix = features.corr().abs()
    upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(np.bool))
    to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.70)]
    print("Dropping these variables because of multicollinearity")
    print("\n")
    print(to_drop)

    features = features.drop(to_drop, axis=1)
    features['Year'] = Year_column['ActiveYear']
    features['Target'] = target['Target']
    
    # split time wise because of time series classlification
    train_set = features[features['Year'] <2017]
    validation_set = features[(features['Year'] >=2017) & (features['Year'] <2019)]
    test_set = features[features['Year'] >2018]
    
    x_train = train_set.drop(['Target', 'Year'], axis = 1)
    y_train = train_set['Target']
    
    x_val = validation_set.drop(['Target', 'Year'], axis = 1)
    y_val = validation_set['Target']
    
    x_test = test_set.drop(['Target', 'Year'], axis = 1)
    y_test = test_set['Target']
    s1 = x_val.shape
    s2 = x_test.shape
    print("\nFeatures Before Prepending {} days of data:".format(T))
    print("Validation set: {}".format(s1))
    print("Test set: {}".format(s2))
    
    prepend_features1 = x_train.iloc[-(T):]
    prepend_features2 = x_val.iloc[-(T):]
    
    x_val = pd.concat([prepend_features1, x_val], axis=0)
    x_test = pd.concat([prepend_features2, x_test], axis=0)
    s3 = x_val.shape
    s4 = x_test.shape
    print("\nFeatures After Prepending {} days of data:".format(T))
    print("Validation set: {}".format(s3))
    print("Test set: {}".format(s4))
    
    x_train, y_train = transformed_features(x_train, y_train, T)
    x_val, y_val = transformed_features(x_val, y_val, T)
    x_test, y_test = transformed_features(x_test, y_test, T)
    s5 = x_train.shape
    s6 = x_val.shape
    s7 = x_test.shape
    
    print("\nFeatures After Transforming and scaling the data:")
    print("train set: {}".format(s5))
    print("Validation set: {}".format(s6))
    print("Test set: {}".format(s7))
    
    return x_train, y_train, x_val, y_val, x_test, y_test    
          
```

#### Functions for Modeling

```python
# Build the Model
def lstm_model(x_train, y_train, x_val, y_val, x_test, y_test, T,N, epoch, size):
    model = Sequential()
    model.add(LSTM(input_shape=(T, N), units=4, activation='tanh', 
                   recurrent_activation='hard_sigmoid', kernel_regularizer=l2(.01), 
                   recurrent_regularizer=l2(.01),
                   return_sequences=True, return_state=False))

    model.add(BatchNormalization())
    model.add(LSTM(units=4, activation='tanh', recurrent_activation='hard_sigmoid', 
                   kernel_regularizer=l2(.01), recurrent_regularizer=l2(.01), 
                   return_sequences=True, return_state=False))

    model.add(BatchNormalization())
    model.add(LSTM(units=4, activation='tanh', recurrent_activation='hard_sigmoid', 
                   kernel_regularizer=l2(0.1), recurrent_regularizer=l2(0.1), 
                   return_sequences=False, return_state=False))
    
    model.add(BatchNormalization())
    model.add(Dense(units=1, activation='sigmoid'))
    # Compile the model with Adam optimizer
    model.compile(loss='binary_crossentropy',
                      metrics=['accuracy'],
                      optimizer=Adam(lr=.001))
    print(model.summary())
    lstm = results2(model, x_train, y_train, x_val, y_val, x_test, y_test, epoch, size)
    return lstm
```

```python
def results2(model, x_train, y_train, x_val, y_val, x_test, y_test, epoch, size):
    
     # Define a learning rate decay method:
    lr_decay = ReduceLROnPlateau(monitor='loss', 
                             patience=1, verbose=0, 
                             factor=0.5, min_lr=1e-8)
    # Define Early Stopping:
    early_stop = EarlyStopping(monitor='val_acc', min_delta=0, 
                               patience=30, verbose=1, mode='auto',
                               baseline=0, restore_best_weights=True)
    
    start = time()
    History = model.fit(x_train, y_train,
                        epochs=epoch,
                        batch_size=size,
                        validation_data=(x_val, y_val),
                        shuffle=True,verbose=0,
                        callbacks=[lr_decay, early_stop])
    print('-'*65)
    print(f'Training was completed in {time() - start:.2f} secs')
    print('-'*65)
    
    print('\n')
    print('Score for Model Testing')
    print(model.evaluate(x_test,y_test))
    
    
    history_dict = History.history
    acc = history_dict['accuracy']
    val_acc = history_dict['val_accuracy']
    loss_values = history_dict['loss']
    val_loss_values = history_dict['val_loss']
    epochs = range(1,len(acc) + 1)

    # Plotting metrics
    plt.plot(epochs, acc,  'bo', label = 'Training accuracy')
    plt.plot(epochs, val_acc, 'b', label = 'Validation accuracy')
    plt.title('Training and Validation Accuracy')
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.figure()
    plt.plot(epochs, loss_values,  'bo', label = 'Training Loss')
    plt.plot(epochs, val_loss_values, 'b', label = 'Validation loss')
    plt.title('Training and Validation Loss')
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.show()

    
    y_preds = model.predict_classes(x_test)
    accuracy = accuracy_score(y_test, y_preds)*100
    print("Accuracy score: {}{}".format(round(accuracy, 2), '%'))
    
    precision = precision_score(y_test, y_preds)*100
    print("Precision score: {}{}".format(round(precision, 2), "%"))

    recall = recall_score(y_test, y_preds)*100
    print("Recall score: {}{}".format(round(recall, 2), "%"))
  
    
    cf_matrix = confusion_matrix(y_test, y_preds)
    
    fig, ax = plt.subplots(figsize = (10,8))
    ax = sns.heatmap(cf_matrix/np.sum(cf_matrix), annot=True, fmt='.2%', cmap='Blues')

    ax.set_title('Confusion Matrix with labels\n\n', loc='left')
    ax.set_xlabel('\nPredicted Values')
    ax.set_ylabel('Actual Values ')
    ax.xaxis.set_ticklabels(['False','True'])
    ax.yaxis.set_ticklabels(['False','True'])
    plt.show()
    
    return model
    
```

### Train Model with Feature Selecttion using ANOVA F-Statistics

```python
# load libraries
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2, f_classif, mutual_info_classif
from functools import partial

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
def FeatureSelector(df, y, score_func, k):
    
    # apply SelectKBest class to extract top features from categorical variables only 
    feature_selector = SelectKBest(score_func=score_func, k=k)
    features_kbest = feature_selector.fit(df,y)

    dfscores = pd.DataFrame(features_kbest.scores_)
    dfcolumns = pd.DataFrame(df.columns)

    #concat two dataframes for better visualization 
    featureScores = pd.concat([dfcolumns,dfscores],axis=1)
    featureScores.columns = ['Features','Score']  #naming the dataframe columns
    top_feature = featureScores.nlargest(k,'Score')  # best features
    return top_feature
```
